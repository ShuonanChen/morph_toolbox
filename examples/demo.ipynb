# morph_toolbox demo

A quick tour using the example neurons in `examples/data/` (MouseLight / Janelia JSON, in CCFv3 microns):

1. **Convert** a JSON neuron description to SWC
2. **Load** the SWC reconstruction
3. **Visualize** it in 2D and 3D

Run top to bottom.

```python
from pathlib import Path

import matplotlib.pyplot as plt

import morph_toolbox as mt

# %matplotlib inline  # uncomment for static inline figures

EXAMPLES = Path.cwd()  # assumes the notebook runs from the examples/ folder
DATA = EXAMPLES / "data"

# Available example neurons (N015 is large and slower to plot).
sorted(p.name for p in DATA.glob("*.json"))
```

## 1. Convert JSON → SWC

These files use the MouseLight format, where the soma, axon, and dendrite are stored as separate arbors. `morph_toolbox` auto-detects that and merges them into one SWC tree (single soma root). `json_to_swc` writes a standard `.swc`; `json_to_morphology` returns the in-memory object directly.

```python
# >>> Pick which neuron to work with <<<
JSON_PATH = DATA / "N038-658735.json"   # try N035-658735 or N015-706301 too
OUT_SWC = JSON_PATH.with_suffix(".swc")

mt.json_to_swc(JSON_PATH, OUT_SWC)
print(f"wrote {OUT_SWC.name}")
```

## 2. Load the SWC file

```python
# (You can also load any .swc you already have: mt.load_swc("my_neuron.swc").
#  Pass reindex=True if the file has gapped, non-consecutive ids.)
morph = mt.load_swc(OUT_SWC)
morph
```

```python
# Headline morphometrics
for key, value in morph.summary().items():
    print(f"{key:>16}: {value}")
```

```python
# The raw SWC node table is just a pandas DataFrame:
morph.nodes.head()
```

## 3. Visualize

Segments are colored by SWC point type (soma, axon, dendrite, ...).

```python
# 2D projection (try 'xy', 'xz', or 'yz')
mt.plot_2d(morph, projection="xy")
plt.show()
```

```python
# All three projections side by side
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for ax, proj in zip(axes, ["xy", "xz", "yz"]):
    mt.plot_2d(morph, projection=proj, ax=ax, legend=False, title=proj)
fig.tight_layout()
plt.show()
```

```python
# 3D view (can take a few seconds for large neurons)
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection="3d")
mt.plot_3d(morph, ax=ax)
plt.show()
```
